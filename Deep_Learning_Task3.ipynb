```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import BatchNormalization, LayerNormalization
```

```python
np.set_printoptions(precision=3, suppress=True)

# ----------------------------------------------------------------------
# 1. Create a small dataset (random input values)
# ----------------------------------------------------------------------
# Shape: (batch_size=4, features=5) -> 4 "samples", each with 5 features
np.random.seed(42)
data = np.random.uniform(low=0, high=100, size=(4, 5)).astype(np.float32)

print("=" * 60)
print("Original Input Values")
print("=" * 60)
print(data)
```

```python
# ----------------------------------------------------------------------
# 2. Apply Batch Normalization
# ----------------------------------------------------------------------
# BatchNorm normalizes each FEATURE (column) using the mean/variance
# computed ACROSS THE BATCH (all 4 samples), for that feature.
bn_layer = BatchNormalization()
# training=True forces it to use the current batch's statistics
# instead of the (untrained) running averages.
bn_output = bn_layer(data, training=True)

print("\n" + "=" * 60)
print("Batch Normalization Output")
print("=" * 60)
print(bn_output.numpy())
```

```python
# ----------------------------------------------------------------------
# 3. Apply Layer Normalization
# ----------------------------------------------------------------------
# LayerNorm normalizes each SAMPLE (row) using the mean/variance
# computed ACROSS THE FEATURES of that single sample.
ln_layer = LayerNormalization()
ln_output = ln_layer(data)

print("\n" + "=" * 60)
print("Layer Normalization Output")
print("=" * 60)
print(ln_output.numpy())
```

```python
# ----------------------------------------------------------------------
# 4. Sanity check: show the stats each method actually normalizes over
# ----------------------------------------------------------------------
print("\n" + "=" * 60)
print("Quick check of normalization axis")
print("=" * 60)
print("Per-FEATURE mean across the batch (BatchNorm's target ~0):")
print(bn_output.numpy().mean(axis=0))
print("\nPer-SAMPLE mean across features (LayerNorm's target ~0):")
print(ln_output.numpy().mean(axis=1))
```

```python
# ----------------------------------------------------------------------
# 5. Explanation
# ----------------------------------------------------------------------
explanation = """
Comparison of Batch Normalization and Layer Normalization
------------------------------------------------------------
- Batch Normalization (BN) normalizes each feature (column) using the
  mean and variance computed ACROSS ALL SAMPLES in the current batch.
  This keeps activations at a stable scale during training, which lets
  the network use higher learning rates and converge faster.
- Layer Normalization (LN) normalizes each sample (row) using the mean
  and variance computed ACROSS THAT SAMPLE'S OWN FEATURES, independent
  of any other sample in the batch.
- Key difference: BN's statistics depend on the batch (so results
  change with batch size and are unstable for batch size = 1), while
  LN's statistics depend only on a single sample (so it works
  identically regardless of batch size) - this is why LN is preferred
  in RNNs/Transformers where batch statistics are less meaningful.
"""
print(explanation)
```
